#Files,Exception Handling ,Logging and Memory Management Questions


1.  What is the difference between interpreted and compiled languages?

      Interpreted languages run line-by-line at runtime (Python).
      Compiled languages are converted to machine code before execution (C, C++).
      Interpreted = slower but flexible; Compiled = faster but needs compilation.

2. What is exception handling in Python?

      It handles runtime errors using try, except, else, finally.
      It prevents program crashes and allows graceful error messages.

3. What is the purpose of the finally block in exception handling?

      finally runs always, whether exception occurs or not.
      Used for cleanup like closing files or releasing resources.

4. What is logging in Python?

      Logging records events, errors, and information during program execution.
      Used for debugging and monitoring applications.

5. What is the significance of the __del__ method in Python?

      __del__ is a destructor called when an object is about to be deleted.
      Used to release resources but not reliable for time-critical cleanup.

6. What is the difference between import and from ... import in Python?

      import module imports whole module → use module.function().
      from module import func imports only specific parts → use func() directly.

7. How can you handle multiple exceptions in Python?

      Use multiple except blocks or tuple of exceptions.

      try:
          x = 1/0
      except (ZeroDivisionError, ValueError):
          print("Handled")

8. What is the purpose of the with statement when handling files in Python?

      It automatically opens and closes the file.
      Prevents resource leakage.

      with open("a.txt") as f:
          print(f.read())

9.  what is Difference between multithreading & multiprocessing?

      Multithreading → multiple threads share same memory (good for I/O).
      Multiprocessing → multiple processes with separate memory (good for CPU tasks).

10. What is the advantages of using looging in a program?

      Helps debug errors, track program flow, and store runtime info.
      Better than print() because it supports levels and file logging.

11. What is memory management in Python?

      Python manages memory automatically using a private heap and garbage collector.

12. What are the basic steps involved in exception handling in Python?

       Write risky code in try, handle error in except, optional else for no-error part, and finally for cleanup.

13. Why memory management is important?

       It avoids memory leaks, improves performance, and ensures efficient resource usage.

14. What is the role of try and except in exception handling?

      try contains code that might cause an error.
      except catches and handles specific exceptions.

15. How does Python's garbage collection system work?

      reference counting + cyclic garbage collector to free unused objects.

16.  What is the purpose of the else block in exception handling?

      else runs only if no exception occurs in try.
      Useful for code that should run only when everything goes right.

17. What are the common logging levels in Python?

     DEBUG, INFO, WARNING, ERROR, CRITICAL.

18.  whta is Difference between os.fork() and multiprocessing?

      os.fork() works only on Unix; directly creates a child process.
      multiprocessing works cross-platform and provides safer, high-level process handling.

19. What is the importance of closing a file in Python?

     It frees memory and ensures data is properly written to disk.

20.  what is Difference between file.read() and file.readline()?

        read() → reads entire file.
        readline() → reads one line at a time.

21.What is the logging module in Python used for?

      It helps write logs to console/files with different levels for debugging and monitoring.

22.What is the os module in Python used for in file handling?

        Provides functions for file/directory operations (create, delete, rename).
        Example: os.remove("file.txt").

23. What are the challenges associated with memory management in Python?

       Circular references, large objects, and fragmentation can cause inefficiency.

24.  How do you raise an exception manually in Python?

        Use raise keyword.

        raise ValueError("Invalid input")

25. Why is it important to use multithreading in certain applications?

      Multithreading is useful for tasks that wait (I/O), like downloading files, so the program doesn't freeze.
      It increases responsiveness and uses CPU time efficiently.

      Example:

      import threading
      def task():
          print("Thread running")
      t = threading.Thread(target=task)
      t.start()

```python
# 6 Write a Python program that logs an error message to a log file when a division by zero exception occurs?
import logging
logging.basicConfig(filename="error.log",level=logging.ERROR)
try:
    a=1/0
except ZeroDivisionError as e:
    logging.error("Handled>>",e)
    logging.shutdown()
    f=open("error.log","r")
    print(f.read())
    f.close()
    print("file closed")



```

```python
# 15.Write a Python program that prints the content of a file and handles the case when the file is empty.
filename = "data.txt"

try:
    with open(filename, "r") as f:
        content = f.read()
        if not content:
            print("File is empty.")
        else:
            print("File content:\n", content)
except FileNotFoundError:
    print("File not found!")
```

```python
# 18. How would you implement a basic logging setup that logs to a file with rotation after 1MB?
import logging
from logging.handlers import RotatingFileHandler

handler = RotatingFileHandler("app.log", maxBytes=1_000_000, backupCount=3)

logging.basicConfig(
    handlers=[handler],
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s"
)

logging.info("This is an informational log.")
logging.error("This is an error log.")
```

```python
# 21.Write a Python program that reads a file and prints the number of occurrences of a specific word.

import re

filename = "story.txt"
word = "python"

try:
    with open(filename, "r") as f:
        text = f.read().lower()
        count = len(re.findall(rf"\b{word.lower()}\b", text))
        print("Occurrences:", count)
except FileNotFoundError:
    print("File not found!")
```

```python
# 22.How can you check if a file is empty before attempting to read its contents?
import os

filename = "test.txt"

if os.path.exists(filename) and os.path.getsize(filename) == 0:
    print("File is empty.")
else:
    print("File has content.")
```

```python
# 23. Write a Python program that writes to a log file when an error occurs during file handling.
import logging

logging.basicConfig(filename="error.log",
                    level=logging.ERROR,
                    format="%(asctime)s - %(levelname)s - %(message)s")

filename = "unknown.txt"

try:
    with open(filename, "r") as f:
        print(f.read())
except Exception as e:
    logging.error(f"Error while reading file: {e}")
    print("An error occurred. Check error.log file.")
```
